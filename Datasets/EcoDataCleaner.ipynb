```python
waterData = pd.read_excel('/kaggle/input/environment-dataset-part-1/Water System Summary_20231215.xlsx')
```

```python
for i in range(len(waterData)):
    waterData.loc[i, 'Zip Code'] = str(waterData.loc[i, 'Zip Code'])#[:5]
```

```python
waterData.sort_values(by=['Zip Code'], inplace = True)
print(waterData)
```

```python
waterData.to_csv("WaterData.csv", encoding='utf-8', index=False)
```

```python
zipData = pd.read_csv('/kaggle/input/environment-dataset-part-3/ZipData.csv')
```

```python
print(zipData)
```

```python
zipData.drop(['state', 'country', 'world_region', 'decommissioned'], axis=1, inplace = True)
```

```python
print(zipData)
```

```python
zipData.rename(columns = {'zip': 'ZIP'}, inplace = True)
print(zipData)
```

```python
merged_data = pd.merge(merged_data, zipData, on='ZIP', how='outer')
print(merged_data)
```

```python
for i in range(len(merged_data)):
    zipStr = str(merged_data.loc[i, 'ZIP'])
    if (zipStr == 'nan'):
        zipStr = ""
    missing = 5 - len(zipStr)
    for j in range(missing):
        zipStr = '0' + zipStr
    merged_data.loc[i, 'ZIP'] = zipStr
    for col in ['no2', 'ozone', 'pm25', 'STATE', 'type', 'primary_city', 'acceptable_cities', 'unacceptable_cities', 'county', 'timezone', 'area_codes', 'latitude', 'longitude', 'irs_estimated_population']:
        dataStr = str(merged_data.loc[i, col])
        if (dataStr == "") or (dataStr == 'nan') or (dataStr == 'NaN'):
            dataStr = "X"
        merged_data.loc[i, col] = dataStr
```

```python
print(merged_data)
```
